```python
import pickle
import pandas as pd
import  warnings
warnings.filterwarnings('ignore')
```

```python
input_data={'X1 transaction date':2013.500,
'X2 house age':13.3,
'X3 distance to the nearest MRT station':561.9845,
'X4 number of convenience stores':5,
'X5 latitude':24.98746,
'X6 longitude':121.54391
}

input_df=pd.DataFrame([input_data])
input_df
```

```python
limits = pickle.load(open("outlier_limits.pkl","rb"))

for col,(lower,upper) in limits.items():
    input_df[col] = input_df[col].clip(lower,upper)

```

```python
input_df
```

```python
scalar=pickle.load(open("scalar.pkl","rb"))
input_df_scaled=scalar.transform(input_df)
input_df_scaled
```

```python
model=pickle.load(open("regr.pkl", "rb"))
```

```python
prediction = model.predict(input_df_scaled)
print(prediction)
```
