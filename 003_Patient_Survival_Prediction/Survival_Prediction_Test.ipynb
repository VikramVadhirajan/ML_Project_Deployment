```python
import pickle
import pandas as pd
import  warnings
warnings.filterwarnings('ignore')
```

```python
input_data={
"N_Days":400,
"Drug":"D-penicillamine",
"Age":21464,
"Sex":"F",
"Ascites":"Y",
"Hepatomegaly":"Y",
"Spiders":"Y",
"Edema":"Y",
"Bilirubin":14.5,
"Cholesterol":261,
"Albumin":2.6,
"Copper":156,
"Alk_Phos":1718,
"SGOT":137.95,
"Tryglicerides":172,
"Platelets":190,
"Prothrombin":12.2,
"Stage":4,
}

input_df=pd.DataFrame([input_data])
input_df
```

```python
input_df.info()
```

```python
cat=[]
num=[]
for i in input_df.columns:
    if input_df[i].dtype=="object":
        cat.append(i)
    else:
        num.append(i)
print(cat) 
print(num) 
```

```python
limits = pickle.load(open("outlier_limits.pkl","rb"))
```

```python
for col,(lower,upper) in limits.items():
    input_df[col] = input_df[col].clip(lower,upper)
```

```python
encoders=pickle.load(open("dep_encoders.pkl","rb"))
for col in cat:
    input_df[col] = encoders[col].transform(input_df[col])
```

```python
input_df
```

```python
zscalar=pickle.load(open("zscaler.pkl","rb"))
input_scaled = zscalar.transform(input_df)
```

```python
LRModel=pickle.load(open("LRModel.pkl","rb"))
prediction = LRModel.predict(input_scaled)
```

```python
print(prediction)
```

```python
indep_encoder=pickle.load(open("indep_encoder.pkl","rb"))
result = indep_encoder.inverse_transform(prediction)
print(result)
```

```python
prediction_proba = LRModel.predict_proba(input_scaled)[0]
print(max(prediction_proba))
```
