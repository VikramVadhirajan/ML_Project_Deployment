```python
import pickle
import pandas as pd
import numpy as np
```

```python
with open('knn_imputer.pkl','rb') as file:
    knn_imputer=pickle.load(file)

with open('simple_imputer.pkl','rb') as file:
    simple_imputer=pickle.load(file)

with open('encoders.pkl','rb') as file:
    encoders=pickle.load(file)

with open('scalar.pkl','rb') as file:
    scaler=pickle.load(file)

with open('random_forest_model.pkl','rb') as file:
    rfmodel=pickle.load(file)
    

    
```

```python
input_data={
"age":48,
"blood_pressure":80,
"specific_gravity":1.02,
"albumin":1,
"sugar":0,
"red_blood_cells":"normal",
"pus_cell":"normal",
"pus_cell_clumps":"notpresent",
"bacteria":"notpresent",
"blood_glucose_random":121,
"blood_urea":36,
"serum_creatinine":1.2,
"sodium":0,
"potassium":0,
"hemoglobin":15.4,
"packed_cell_volume":44,
"white_blood_cell_count":7800,
"red_blood_cell_count":5.2,
"hypertension":"yes",
"diabetes_mellitus":"yes",
"coronary_artery_disease":"no",
"appetite":"good",
"pedal_edema":"no",
"anemia":"no",
}
```

```python
input_df=pd.DataFrame([input_data])
input_df
```

```python
cat=[]
num=[]
for i in input_df.columns:
    if input_df[i].dtype=="object":
        cat.append(i)
    else:
        num.append(i)
print(cat) 
print(num) 
```

```python
input_df[cat] = simple_imputer.fit_transform(input_df[cat])
```

```python
input_df[num] = knn_imputer.fit_transform(input_df[num])
```

```python
input_df
```

```python
for col in cat:
    input_df[col] = encoders[col].transform(input_df[col])
```

```python
input_df
```

```python
## Scaling the input data
input_scaled=scaler.transform(input_df)
input_scaled
```

```python
## PRedict churn
prediction=rfmodel.predict(input_scaled)
prediction
```

```python
prediction_proba = rfmodel.predict_proba(input_scaled)[0][1]
prediction_proba
```
